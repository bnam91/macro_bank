```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

# ... (이전 코드는 그대로 유지)

def wait_for_element(driver, by, value, timeout=30):
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((by, value))
        )
        return element
    except TimeoutException:
        print(f"Element {value} not found within {timeout} seconds")
        return None

def main():
    refresh_count = 0
    max_refresh = 3

    while refresh_count < max_refresh:
        try:
            # Selenium 설정 및 웹페이지 열기
            url = 'https://www.kebhana.com/common/login.do'
            options = Options()
            options.add_experimental_option("detach", True)
            options.add_argument("--disable-blink-features=AutomationControlled")

            driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
            driver.get(url)

            # 페이지 로드 대기
            cert_login_button = wait_for_element(driver, By.CSS_SELECTOR, "#certLogin")
            if not cert_login_button:
                raise TimeoutException("Failed to load the login page")

            print(cert_login_button.text)
            cert_login_button.click()

            # PyAutoGUI를 이용한 인증서 로그인
            def login_with_certificate():
                localdisk_location = wait_for_image(COORDS["localdisktab"], timeout=30)
                if not localdisk_location:
                    raise TimeoutException("Failed to find localdisktab")
                pyautogui.click(localdisk_location)
                tm.sleep(2)

                outdisk_location = wait_for_image(COORDS["outdisktab"], timeout=30)
                if not outdisk_location:
                    raise TimeoutException("Failed to find outdisktab")
                pyautogui.click(outdisk_location)
                tm.sleep(3)
        
                for _ in range(6):   # 탭 6번
                    pyautogui.press('tab')
                    tm.sleep(0.1)
        
                # 비밀번호 입력 (한글자씩 입력)
                password = "!gusqls120"
                for char in password:
                    pyautogui.press(char)
                    tm.sleep(0.1)
        
                commonlogin_location = wait_for_image(COORDS["commonlogin"], timeout=30)
                if not commonlogin_location:
                    raise TimeoutException("Failed to find commonlogin button")
                pyautogui.click(commonlogin_location)

            login_with_certificate()
            tm.sleep(8)

            # 프레임 전환
            frame_id = "hanaMainframe"
            frame = wait_for_element(driver, By.ID, frame_id)
            if not frame:
                raise TimeoutException("Failed to find main frame")
            driver.switch_to.frame(frame)

            # 이체 탭 클릭
            send_tab_location = wait_for_image(COORDS["send_tab"], timeout=30)
            if not send_tab_location:
                raise TimeoutException("Failed to find send tab")
            pyautogui.click(send_tab_location)
            tm.sleep(1)

            # 다계좌 이체 버튼 클릭
            for _ in range(3):   # 탭 3번
                pyautogui.press('tab')

            pyautogui.press('enter')   
            tm.sleep(3)

            # 여기에 나머지 코드를 추가 (스크롤, 데이터 처리, 입력 등)

            break  # 성공적으로 완료되면 루프 종료

        except (TimeoutException, NoSuchElementException) as e:
            print(f"Error occurred: {e}")
            refresh_count += 1
            if refresh_count < max_refresh:
                print(f"Refreshing page. Attempt {refresh_count + 1} of {max_refresh}")
                driver.refresh()
            else:
                print("Max refresh attempts reached. Exiting...")
                break

        finally:
            if 'driver' in locals():
                driver.quit()

if __name__ == "__main__":
    main()
```
